```python
import os
from gtts import gTTS
from IPython.display import Audio
from googletrans import Translator
import IPython.display as ipd
```

**Read Text File**

```python
with open("/content/text.txt", "r") as f:
    text = f.read().replace("\n", "")
```

**Arabic**

```python
tts = gTTS(text, lang="ar",slow=False)
tts.save("artext.mp3")
```

```python
audio_path="entext.mp3"
ipd.Audio(audio_path, autoplay=True)
```

**English**

```python
translator = Translator()
tts = gTTS(text=translator.translate(text,dest="en").text,lang="en",tld='co.uk',slow=False)
tts.save("entext.mp3")
```

```python
audio_path="entext.mp3"
ipd.Audio(audio_path, autoplay=True)
```

**French**

```python
translator = Translator()
tts = gTTS(text=translator.translate(text,dest="fr").text,lang="fr",tld='ca',slow=False)
tts.save("frtext.mp3")
```

```python
audio_path="frtext.mp3"
ipd.Audio(audio_path, autoplay=True)
```
